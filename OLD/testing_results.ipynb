```python
import beam_search_module as bsm
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('stock_data_all_features.csv')
data = data.drop(columns=['dates', 'target'])
data['growth_target'] = data['growth_target'].apply(eval)
data
```

```python
domain_attributes = True
time_series_attributes = True

if not domain_attributes and time_series_attributes:
    data = data.iloc[:, 12:]

if domain_attributes and not time_series_attributes:
    domain_features = data.columns[:12]
    columns_keep = list(domain_features) + ['growth_target']
    data = data[columns_keep]

if domain_attributes and time_series_attributes:
    data = data


omega = data.values.tolist()
column_names = list(data.columns)
beam_width = 20
beam_depth = 3
nr_bins = 4
nr_saved = 10
subgroup_size = 0.05*len(data)
target = 'growth_target'
target_ind = column_names.index(target)
all_time_series = [i[target_ind] for i in omega]
all_time_series = np.array(all_time_series)
print(all_time_series.shape)
targets_baseline = np.mean(all_time_series, axis=0)
print(targets_baseline.shape)
att_indices = list(range(0, len(column_names)))
att_indices.remove(target_ind)
att_columns = [column_names[i] for i in att_indices]
types = bsm.categorize_columns_in_order(data, att_columns)
window_size = 20
```

```python
for row in omega:
    new_row = row[:]
    print(row)
    print(row[target_ind])
    print(window_size)
    break
```

```python
result = bsm.beam_search_with_constraint(omega, targets_baseline, column_names, beam_width, beam_depth, nr_bins, nr_saved, subgroup_size, target, types, window_size, min_quality_improvement=5)
```

```python
print(result)
```

```python
stock_data = data.copy()

for i in range(10):

    quality, description, x = list(result)[i]

    for descriptor in list(description):
        feature, value, func = descriptor
        stock_data = stock_data[func(stock_data[feature], value)]

    print(len(stock_data), quality)
```
